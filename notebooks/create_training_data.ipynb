# Creating a single training data source

Dataset: [Credit Card Fraud Detection- Kaggle](https://www.kaggle.com/competitions/nus-fintech-recruitment/data)

## Importing modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from warnings import filterwarnings
```

## Data paths

```python
train_path = '../Data/train.csv'
test_path = '../Data/test.csv'
customer_path = '../Data/customer.csv'
terminal_path = '../Data/terminal.csv'
```

## Checking each data source file

```python
train = pd.read_csv(train_path)
train.head()
```

```python
customer = pd.read_csv(customer_path)
customer.head()
```

```python
terminal = pd.read_csv(terminal_path)
terminal.head()
```

```python
test = pd.read_csv(test_path)
test.head()
```

## Creating training data

### Droping aviailable terminals from customer data

```python
customer.drop(['available_terminals'], axis=1, inplace=True)
customer.head(2)
```

### Merging train and customer data

```python
train = pd.merge(train, customer, on='CUSTOMER_ID', how='left')
train.head(2)
```

### Merging train and terminal data

```python
train_df = pd.merge(train, terminal, on='TERMINAL_ID', how='left')
```

## Final data set

```python
train_df.head()
```

```python
train_df.to_csv('../Data/final_train.csv', index=False)
```
